```python
y_columns = list(set(train2.columns) - set(X_columns) - set(['id']))
y_columns.sort()
y_columns
```

```python
train3 = train2.loc[ :,  X_columns + y_columns  ]
train3.head(3)
```

```python
temp_li = [] # column별 데이터 묶기

temp_corr_li = [] # column별 데이터 corr 묶기

for i in range(0,36, 5):
    temp_df = train3.iloc[: , i : i+5 ]
    #print(temp_df.columns[0][:-3])
    #print()
    temp_li.append( temp_df )
    temp_corr_li.append( temp_df.corr() )
    #print()
```
